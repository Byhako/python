```python
import seaborn as sns
```

```python
propinas = sns.load_dataset('tips')
propinas.head()
```

```python
propinas.shape
```

```python
sns.distplot(propinas['total_bill'], bins=20);
```

```python
sns.jointplot(data=propinas, x='total_bill', y='tip');
```

```python
sns.jointplot(data=propinas, x='total_bill', y='tip', kind='reg');
```

```python
sns.jointplot(data=propinas, x='total_bill', y='tip', kind='kde');
```

```python
import matplotlib.pyplot as plt
```

```python
flores = sns.load_dataset('iris')
flores.head()
```

```python
grafico = sns.PairGrid(flores)
#grafico.map(plt.scatter)
grafico.map_diag(sns.distplot)
grafico.map_upper(plt.scatter)
grafico.map_lower(sns.kdeplot)
```

```python
grafico2 = sns.FacetGrid(data=propinas, col='time', row='smoker');
grafico2.map(sns.distplot, 'total_bill');
```

```python
grafico2 = sns.FacetGrid(data=propinas, col='time', row='smoker');
grafico2.map(plt.scatter, 'total_bill', 'tip');
```
