* Algoritmo no supervisado para resolver clusterizaci√≥n
* Parte conjunto de **n** observables en **k** grupos
* Se usa en mineria de datos

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
sns.set_style('darkgrid')
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4)
```

```python
plt.scatter(data[0][:,0], data[0][:,1]);
```

```python
from sklearn.cluster import KMeans
```

```python
modelo = KMeans(n_clusters=4)
modelo.fit(data[0])
```

```python
modelo.cluster_centers_
```

Aqui vemos las coordenadas de los centros para cada grupo.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))

ax1.scatter(data[0][:,0], data[0][:,1], c=modelo.labels_, cmap='viridis')
ax1.set_title('Algoritmo k-medias')

ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='viridis')
ax2.set_title('Datos Originales')
```
