```python
import requests
import pandas as pd
from bs4 import BeautifulSoup 
```

```python
data = requests.get('http://www.dinero.com/edicion-impresa/caratula/articulo/los-100-mejores-colegios-medellin-antioquia/165163')
print(data) # si responde 404 es porque no encontro la direccion, si es 200 si la encuentra
```

```python
soup = BeautifulSoup(data.content)
```

```python
tablas = soup.findAll('div', {'class':'theiaStickySidebar'})
tabla = tablas[0].findAll('table')
len(tabla)
```

```python
filas = tabla[0].findAll('tr')
col = filas[0].findAll('td')
len(col)
```

```python
nombresCol = []
for nombre in col:
    nombresCol.append(nombre.text)
nombresCol
```

```python
filas.pop(0)
```

```python
data_table = []
for fila in filas:
    current_data = {}
    i = 0
    for columna in fila.findAll('td'):
        current_data[nombresCol[i]] = columna.text
        i += 1
    data_table.append(current_data)
lista = pd.DataFrame(data_table)
lista
```

```python
lista.to_csv('listaDeColegios.csv')
```

```python
# Filtramos los colegios que estan en medellin
medellin = lista[lista['\nCIUDAD \n']=='\nMEDELLIN \n']
medellin
```
