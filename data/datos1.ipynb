Ahora vamos a realizar una funcion que grafique la columna problematica filtrada con la dependencia que escojamos.

```python
def graficaValoresProblematica(dataframe, dependencia):
    dataframeValores = dataframe[dataframe['Dependencia']==dependencia]
    pd.value_counts(dataframeValores['problematica']).plot(kind='bar')
    return 'success'
```

```python
graficaValoresProblematica(survey, 'CENDI I')
```
