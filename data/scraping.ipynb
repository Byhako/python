```python
import requests
import pandas as pd
from bs4 import BeautifulSoup 
```

```python
data = requests.get('https://es.wikipedia.org/wiki/Anexo:Consolas_de_Nintendo')
print(data) # si responde 404 es porque no encontro la direccion, si es 200 si la encuentra
```

```python
data.content # muestra el html de la pagina
```

```python
soup = BeautifulSoup(data.content)
```

<h4>Ahora vamos a traer la tabla que aparece en la wikipedia, para ello inspeccionamos la tabla
con el navegador para ver como se llama la clase de la tabla en el html y usamos el 
metodo "findAll"

```python
tablas = soup.findAll('table', {'class':'wikitable'})
len(tablas)
```

```python
# tablas[0]
```

```python
filas = tablas[0].findAll('tr')
len(filas)
```

```python
filas[0]
```

```python
col = filas[0].findAll('th')
len(col)
```

```python
nombresCol = []
for nombre in col:
    nombresCol.append(nombre.text)
nombresCol
```

```python
filas.pop(0) # elimino el primero valor de las filas, para quitar los nombres.
```

```python
data_table = []
for fila in filas:
    current_data = {}
    i = 0
    for columna in fila.findAll('td'):
        current_data[nombresCol[i]] = columna.text
        i += 1
    data_table.append(current_data)
pd.DataFrame(data_table)
```

<h2><strong> Traigamos otra tabla de la Wiki <br\>

```python
datos = requests.get('https://es.wikipedia.org/wiki/G8')
print(datos)
sopa = BeautifulSoup(datos.content)
cuadros = sopa.findAll('table', {'class':'wikitable'})
filas = cuadros[0].findAll('tr')
col = filas[0].findAll('th')

# Obtengo los nombres de las columnas
namesCol = []
for nombre in col:
    namesCol.append(nombre.text)
namesCol.pop()

# Elimino el primero valor de las filas, para quitar los nombres.
filas.pop(0) 

# Generamos una lista de diccionarios que contienen los valores de las columnas para cada fila.
data_table = []
for fila in filas:
    current_data = {}
    i = 0
    for columna in fila.findAll('td'):
        current_data[namesCol[i]] = columna.text
        i += 1
        if i == 5: break
    data_table.append(current_data)

# Con pandas transformamos la lista a una tabla.
pd.DataFrame(data_table)
```
